```python
import pandas as pd
from rdkit import Chem
```

```python
df = pd.read_csv('newdata/Maybridge_HitDiscover.csv')
```

```python
df
```

```python
sel = ['SMILES', 'Code']
df1 = df[sel]
df1
```

```python
def to_canonical(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return Chem.MolToSmiles(mol, isomericSmiles=True)
    else:
        return None
```

```python
smiles_list = df1['SMILES']

canonical_smiles_list = [to_canonical(smiles) for smiles in smiles_list]

df1.loc[:, 'SMILES'] = canonical_smiles_list
```

```python
df1
```

```python
df1 = df1[df1.SMILES.notna()]
df1
```

```python
df1.to_csv('maybridge/Maybridge.smi', sep='\t', index=False, header=False)
```
