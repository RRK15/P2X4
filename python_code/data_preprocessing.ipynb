## Data from ChEMBL database

```python
df1 = pd.read_csv('data/chembl/p2x4_chembl_data.csv')
df1
```

## Data from BindingDB

```python
df2 = pd.read_csv('data/bindingdb/p2x4.tsv', sep='\t')
df2
```

```python
filtered_df = df2[df2['Target Source Organism According to Curator or DataSource'] == 'Homo sapiens']
filtered_df
```

```python
selection = ['BindingDB Reactant_set_id','Ligand SMILES','IC50 (nM)']
df3 = filtered_df[selection]
df3
```

```python
df1.columns = ['X', 'Y', 'Z']
df3.columns = ['X', 'Y', 'Z']
```

```python
df3 = df3[df3.Y.notna()]
df3 = df3[df3.Z.notna()]
```

```python
df3['Z'] = df3['Z'].str.replace('[<>]', '')
df3
```

```python
def to_canonical(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return Chem.MolToSmiles(mol, isomericSmiles=True)
    else:
        return None
```

```python
smiles_list = df3['Y']

canonical_smiles_list = [to_canonical(smiles) for smiles in smiles_list]

df3.loc[:, 'Y'] = canonical_smiles_list
```

```python
df3
```

## Data from PubChem database

```python
df4 = pd.read_csv('data/pubchem/pubchem_smiles.csv')
df4.columns = ['X', 'Y', 'Z']
df4
```

```python
# Concatenate the data frames vertically
merged_df = pd.concat([df1, df3, df4])

# Reset index (optional)
merged_df = merged_df.reset_index(drop=True)

# Output the merged data frame
print(merged_df)
```

```python
len(merged_df.Y.unique())
```

```python
duplicate_df = merged_df[merged_df.duplicated(subset='Y', keep=False)]

# Group by column Y and aggregate columns X and Z
new_df = duplicate_df.groupby('Y').agg({'Y': 'first', 'X': lambda x: list(x), 'Z': lambda z: list(z)})
new_df
```

```python
df = merged_df.drop_duplicates(['Y']).reset_index(drop=True)
df
```

```python
df.columns = ['IDs', 'canonical_smiles', 'IC50']
```

```python
df.to_csv('data/merged_data.csv', index=False)
```

## change IC50 to pIC50

```python
df.IC50.describe()
```

```python
df['IC50'] = df['IC50'].astype(float)
```

```python
df.IC50.describe()
```

```python
def norm_value(input):
    norm = []

    for i in input['IC50']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['IC50_norm'] = norm
    x = input.drop('IC50', 1)
        
    return x
```

```python
df_norm = norm_value(df)
df_norm
```

```python
df_norm.IC50_norm.describe()
```

```python
def pIC50(input):
    pIC50 = []

    for i in input['IC50_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('IC50_norm', 1)
        
    return x
```

```python
df = pIC50(df_norm)
df
```

## checking charge of the compounds

```python
charges = []
for smi in df['canonical_smiles']:
    mol = Chem.MolFromSmiles(smi)
    charge = sum(atom.GetFormalCharge() for atom in mol.GetAtoms())
    charges.append(charge)

# Add the charges as a new column to the DataFrame
df['charge'] = charges
```
